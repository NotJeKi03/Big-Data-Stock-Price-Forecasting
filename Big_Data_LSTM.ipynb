Prediksi 5 Bulan kedepan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error

# Membaca data
df = pd.read_csv("tan.us.txt")

# Mengkonversi kolom 'Date' menjadi datetime
df['Date'] = pd.to_datetime(df['Date'])

# Mengatur 'Date' sebagai index
df.set_index('Date', inplace=True)

# Menggunakan kolom 'Close' untuk prediksi harga
data = df[['Close']]

# Normalisasi data menggunakan MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# Membuat data untuk LSTM dengan menggunakan sliding window
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Membagi data menjadi training dan testing
time_step = 10  # Jumlah hari untuk memprediksi hari berikutnya
X, y = create_dataset(data_scaled, time_step)

# Membagi data menjadi training dan testing (80% untuk training dan 20% untuk testing)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Mengubah bentuk data agar sesuai dengan input LSTM [samples, time_steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Membuat model LSTM
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=1))

# Kompilasi model
model.compile(optimizer='adam', loss='mean_squared_error')

# Melatih model
model.fit(X_train, y_train, epochs=20, batch_size=32)

# Melakukan prediksi pada data uji
predictions = model.predict(X_test)

# Membalikkan normalisasi untuk mendapatkan nilai harga yang asli
predictions_rescaled = scaler.inverse_transform(predictions)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Menghitung Mean Squared Error (MSE)
mse = mean_squared_error(y_test_rescaled, predictions_rescaled)
print(f"Mean Squared Error: {mse}")

# Visualisasi hasil prediksi vs aktual
plt.figure(figsize=(12,7))
plt.plot(y_test_rescaled, color='red', label='Actual Price')
plt.plot(predictions_rescaled, color='blue', label='Predicted Price')
plt.title('Stock Price Prediction with LSTM')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Untuk memprediksi harga 5 bulan ke depan (misalnya 100 hari kerja)
forecast_steps = 100  # 5 minggu ke depan (sekitar 100 hari kerja)
last_data = data_scaled[-time_step:]  # Mengambil data terakhir untuk prediksi
last_data = last_data.reshape(1, time_step, 1)

# Menyimpan hasil prediksi untuk langkah-langkah ke depan
forecast_rescaled = []

# Melakukan iterasi untuk prediksi langkah demi langkah (iteratif)
for _ in range(forecast_steps):
    # Melakukan prediksi untuk langkah berikutnya
    next_step = model.predict(last_data)
    forecast_rescaled.append(next_step[0, 0])  # Menyimpan hasil prediksi

    # Memperbarui data untuk langkah berikutnya
    last_data = np.append(last_data[:, 1:, :], next_step.reshape(1, 1, 1), axis=1)

# Membalikkan normalisasi prediksi
forecast_rescaled = np.array(forecast_rescaled)
forecast_rescaled = scaler.inverse_transform(forecast_rescaled.reshape(-1, 1))

# Membuat tanggal prediksi
forecast_dates = pd.date_range(df.index[-1], periods=forecast_steps + 1, freq='B')[1:]

# Visualisasi hasil prediksi 5 bulan ke depan
plt.figure(figsize=(12,7))
plt.plot(df.index, data['Close'], color='blue', label='Historical Prices')
plt.plot(forecast_dates, forecast_rescaled, color='red', linestyle='dashed', label='Predicted Prices (Next 5 months)')
plt.title('Stock Price Forecast for the Next 5 Weeks using LSTM')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Fungsi untuk membuat dataset dengan sliding window
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Fungsi untuk membuat model LSTM
def build_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dense(units=1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Load data untuk lima saham
stock_files = ['tan.us.txt', 'tao.us.txt', 'tbt.us.txt', 'tbf.us.txt']
stocks_data = {}

for file in stock_files:
    df = pd.read_csv(file)
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    stocks_data[file] = df[['Close']]  # Menggunakan kolom 'Close'

# Normalisasi dan persiapan data
scalers = {}
X_train_list, X_test_list, y_train_list, y_test_list, predictions_list = [], [], [], [], []

for stock, data in stocks_data.items():
    scaler = MinMaxScaler(feature_range=(0, 1))
    data_scaled = scaler.fit_transform(data)
    scalers[stock] = scaler

    # Membuat dataset
    time_step = 10
    X, y = create_dataset(data_scaled, time_step)

    # Membagi data menjadi training dan testing
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Reshape data untuk LSTM
    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    # Menyimpan data untuk training dan testing
    X_train_list.append(X_train)
    X_test_list.append(X_test)
    y_train_list.append(y_train)
    y_test_list.append(y_test)

    # Membuat dan melatih model
    model = build_model((X_train.shape[1], 1))
    model.fit(X_train, y_train, epochs=20, batch_size=32)

    # Melakukan prediksi
    predictions = model.predict(X_test)
    predictions_rescaled = scaler.inverse_transform(predictions)
    y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Menyimpan hasil prediksi
    predictions_list.append(predictions_rescaled)

    # Menghitung MSE dan R-squared
    mse = mean_squared_error(y_test_rescaled, predictions_rescaled)
    r2 = r2_score(y_test_rescaled, predictions_rescaled)

    print(f"Stock: {stock}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"R-squared (R²): {r2:.4f}")

# Menentukan periode waktu yang sama untuk menghitung korelasi
min_length = min(len(predictions) for predictions in predictions_list)

# Memotong semua prediksi agar memiliki panjang yang sama
predictions_list_trimmed = [pred[:min_length] for pred in predictions_list]

# Menghitung korelasi antara saham
correlations = []
for i in range(len(stock_files)):
    for j in range(i + 1, len(stock_files)):
        correlation = np.corrcoef(predictions_list_trimmed[i].flatten(), predictions_list_trimmed[j].flatten())[0, 1]
        correlations.append((stock_files[i], stock_files[j], correlation))

# Menampilkan hasil korelasi
print("\nKorelasi antar saham:")
for stock1, stock2, corr in correlations:
    print(f"Korelasi antara {stock1} dan {stock2}: {corr:.4f}")


# Visualisasi tren gabungan
plt.figure(figsize=(14, 7))
for i, stock in enumerate(stock_files):
    plt.plot(stocks_data[stock].index, stocks_data[stock].values, label=f'Actual {stock}', linestyle='-')
    plt.plot(stocks_data[stock].index[-len(predictions_list[i]):], predictions_list[i], label=f'Predicted {stock}', linestyle='--')

plt.title('Comparison of Actual and Predicted Prices for 4 Stocks')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
