Prediksi 5 Bulan kedepan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error

# Membaca data
df = pd.read_csv("tan.us.txt")

# Mengkonversi kolom 'Date' menjadi datetime
df['Date'] = pd.to_datetime(df['Date'])

# Mengatur 'Date' sebagai index
df.set_index('Date', inplace=True)

# Menggunakan kolom 'Close' untuk prediksi harga
data = df[['Close']]

# Normalisasi data menggunakan MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# Membuat data untuk LSTM dengan menggunakan sliding window
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Membagi data menjadi training dan testing
time_step = 10  # Jumlah hari untuk memprediksi hari berikutnya
X, y = create_dataset(data_scaled, time_step)

# Membagi data menjadi training dan testing (80% untuk training dan 20% untuk testing)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Mengubah bentuk data agar sesuai dengan input LSTM [samples, time_steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Membuat model LSTM
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=1))

# Kompilasi model
model.compile(optimizer='adam', loss='mean_squared_error')

# Melatih model
model.fit(X_train, y_train, epochs=20, batch_size=32)

# Melakukan prediksi pada data uji
predictions = model.predict(X_test)

# Membalikkan normalisasi untuk mendapatkan nilai harga yang asli
predictions_rescaled = scaler.inverse_transform(predictions)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Menghitung Mean Squared Error (MSE)
mse = mean_squared_error(y_test_rescaled, predictions_rescaled)
print(f"Mean Squared Error: {mse}")

# Visualisasi hasil prediksi vs aktual
plt.figure(figsize=(12,7))
plt.plot(y_test_rescaled, color='red', label='Actual Price')
plt.plot(predictions_rescaled, color='blue', label='Predicted Price')
plt.title('Stock Price Prediction with LSTM')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Untuk memprediksi harga 5 bulan ke depan (misalnya 100 hari kerja)
forecast_steps = 100  # 5 minggu ke depan (sekitar 100 hari kerja)
last_data = data_scaled[-time_step:]  # Mengambil data terakhir untuk prediksi
last_data = last_data.reshape(1, time_step, 1)

# Menyimpan hasil prediksi untuk langkah-langkah ke depan
forecast_rescaled = []

# Melakukan iterasi untuk prediksi langkah demi langkah (iteratif)
for _ in range(forecast_steps):
    # Melakukan prediksi untuk langkah berikutnya
    next_step = model.predict(last_data)
    forecast_rescaled.append(next_step[0, 0])  # Menyimpan hasil prediksi

    # Memperbarui data untuk langkah berikutnya
    last_data = np.append(last_data[:, 1:, :], next_step.reshape(1, 1, 1), axis=1)

# Membalikkan normalisasi prediksi
forecast_rescaled = np.array(forecast_rescaled)
forecast_rescaled = scaler.inverse_transform(forecast_rescaled.reshape(-1, 1))

# Membuat tanggal prediksi
forecast_dates = pd.date_range(df.index[-1], periods=forecast_steps + 1, freq='B')[1:]

# Visualisasi hasil prediksi 5 bulan ke depan
plt.figure(figsize=(12,7))
plt.plot(df.index, data['Close'], color='blue', label='Historical Prices')
plt.plot(forecast_dates, forecast_rescaled, color='red', linestyle='dashed', label='Predicted Prices (Next 5 months)')
plt.title('Stock Price Forecast for the Next 5 Weeks using LSTM')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()


# Membuat DataFrame untuk hasil forecasting
forecast_results = pd.DataFrame({
    'Date': forecast_dates,  # Tanggal hasil prediksi
    'Forecasted Price': forecast_rescaled.flatten()  # Harga hasil prediksi
})

# Menampilkan 10 data pertama dari hasil forecasting
print(forecast_results.head(10))

# Menyimpan hasil ke dalam file CSV (opsional)
forecast_results.to_csv("forecast_results.csv", index=False)

# Visualisasi hasil forecast untuk memastikan kebenaran
plt.figure(figsize=(12, 7))
plt.plot(df.index, data['Close'], color='blue', label='Historical Prices')
plt.plot(forecast_results['Date'], forecast_results['Forecasted Price'], color='red', linestyle='dashed', label='Forecasted Prices (Next Steps)')
plt.title('Forecasted Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error

# Membaca data
df = pd.read_csv("tbt.us.txt")

# Mengkonversi kolom 'Date' menjadi datetime
df['Date'] = pd.to_datetime(df['Date'])

# Mengatur 'Date' sebagai index
df.set_index('Date', inplace=True)

# Menggunakan kolom 'Close' untuk prediksi harga
data = df[['Close']]

# Normalisasi data menggunakan MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# Membuat data untuk LSTM dengan menggunakan sliding window
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Membagi data menjadi training dan testing
time_step = 10  # Jumlah hari untuk memprediksi hari berikutnya
X, y = create_dataset(data_scaled, time_step)

# Membagi data menjadi training dan testing (80% untuk training dan 20% untuk testing)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Mengubah bentuk data agar sesuai dengan input LSTM [samples, time_steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Membuat model LSTM
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=1))

# Kompilasi model
model.compile(optimizer='adam', loss='mean_squared_error')

# Melatih model
model.fit(X_train, y_train, epochs=20, batch_size=32)

# Melakukan prediksi pada data uji
predictions = model.predict(X_test)

# Membalikkan normalisasi untuk mendapatkan nilai harga yang asli
predictions_rescaled = scaler.inverse_transform(predictions)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# Menghitung Mean Squared Error (MSE)
mse = mean_squared_error(y_test_rescaled, predictions_rescaled)
print(f"Mean Squared Error: {mse}")

# Visualisasi hasil prediksi vs aktual
plt.figure(figsize=(12,7))
plt.plot(y_test_rescaled, color='red', label='Actual Price')
plt.plot(predictions_rescaled, color='blue', label='Predicted Price')
plt.title('Stock Price Prediction with LSTM')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Untuk memprediksi harga 5 bulan ke depan (misalnya 100 hari kerja)
forecast_steps = 100  # 5 minggu ke depan (sekitar 100 hari kerja)
last_data = data_scaled[-time_step:]  # Mengambil data terakhir untuk prediksi
last_data = last_data.reshape(1, time_step, 1)

# Menyimpan hasil prediksi untuk langkah-langkah ke depan
forecast_rescaled = []

# Melakukan iterasi untuk prediksi langkah demi langkah (iteratif)
for _ in range(forecast_steps):
    # Melakukan prediksi untuk langkah berikutnya
    next_step = model.predict(last_data)
    forecast_rescaled.append(next_step[0, 0])  # Menyimpan hasil prediksi

    # Memperbarui data untuk langkah berikutnya
    last_data = np.append(last_data[:, 1:, :], next_step.reshape(1, 1, 1), axis=1)

# Membalikkan normalisasi prediksi
forecast_rescaled = np.array(forecast_rescaled)
forecast_rescaled = scaler.inverse_transform(forecast_rescaled.reshape(-1, 1))

# Membuat tanggal prediksi
forecast_dates = pd.date_range(df.index[-1], periods=forecast_steps + 1, freq='B')[1:]

# Visualisasi hasil prediksi 5 bulan ke depan
plt.figure(figsize=(12,7))
plt.plot(df.index, data['Close'], color='blue', label='Historical Prices')
plt.plot(forecast_dates, forecast_rescaled, color='red', linestyle='dashed', label='Predicted Prices (Next 5 months)')
plt.title('Stock Price Forecast for the Next 5 Weeks using LSTM')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()


# Membuat DataFrame untuk hasil forecasting
forecast_results = pd.DataFrame({
    'Date': forecast_dates,  # Tanggal hasil prediksi
    'Forecasted Price': forecast_rescaled.flatten()  # Harga hasil prediksi
})

# Menampilkan 10 data pertama dari hasil forecasting
print(forecast_results.head(10))

# Menyimpan hasil ke dalam file CSV (opsional)
forecast_results.to_csv("forecast_results.csv", index=False)

# Visualisasi hasil forecast untuk memastikan kebenaran
plt.figure(figsize=(12, 7))
plt.plot(df.index, data['Close'], color='blue', label='Historical Prices')
plt.plot(forecast_results['Date'], forecast_results['Forecasted Price'], color='red', linestyle='dashed', label='Forecasted Prices (Next Steps)')
plt.title('Forecasted Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
